```python
Comments=[]
Comment_upvote=[]
Time_posted=[] 
```

```python
comments_tags=driver.find_elements_by_xpath('//yt-formatted-string[@class="style-scope ytd-comment-renderer"]')
len(comments_tags)
```

```python
comments=[]
cnt=0
for i in comments_tags:
    if cnt%2!=0:
        comments.append(i.text)
    cnt+=1
```

```python
comments
```

```python
upvotes_tags=driver.find_elements_by_xpath('//span[@class="style-scope ytd-comment-action-buttons-renderer"]')
len(upvotes_tags)
```

```python
comments_upvote=[]
cnt=0
for i in upvotes_tags:
    if cnt%2!=0:
        comments_upvote.append(i.text)
    cnt+=1
```

```python
comments_upvote
```

```python
time_tags=driver.find_elements_by_xpath('//yt-formatted-string[@class="published-time-text above-comment style-scope ytd-comment-renderer"]')
len(time_tags)
```

```python
Time_posted=[]
for i in time_tags:
    Time_posted.append(i.text)
```
