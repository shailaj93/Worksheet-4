```python
dist_tags=driver.find_elements_by_xpath('//span[@class="description"]')
len(dist_tags)
```

```python
distance_from_city_centre=[]
for i in dist_tags:
    distance_from_city_centre.append(i.text)
```

```python
dorm_tags=driver.find_elements_by_xpath('//div[@class="prices-col"]/a[1]/div[2]')
len(dorm_tags)
```

```python
dorms_from_price=[] 
for i in dorm_tags:
    dorms_from_price.append(i.text)
dorms_from_price
```

```python
private_tags=driver.find_elements_by_xpath('//div[@class="prices-col"]/a[1]/div[1]')
len(private_tags)
```

```python
privates_from_price=[] 
for i in private_tags:
    privates_from_price.append(i.text)
privates_from_price
```

```python
Url=[]
for i in driver.find_elements_by_xpath("//h2[@class='title title-6']/a"):
    Url.append(i.get_attribute("href"))
```

```python
len(Url)
```

```python
Url
```

```python
hostel_name=[]
ratings=[]
total_reviews=[] 
overall_reviews=[] 
facilities=[] 
property_description=[]
```

```python
for i in Url:
    driver.get(i)
    time.sleep(10)
    
    try:
        name=driver.find_element_by_xpath('//div[@class="title-2"]')
        hostel_name.append(name.text.replace("/n","new line"))
    except:
        hostel_name.append('-')
        
    try:
        name=driver.find_element_by_xpath('//div[@class="score orange big"]')
        ratings.append(name.text.replace("/n","new line"))
    except:
        ratings.append('-')
    
    try:
        name=driver.find_element_by_xpath('//div[@class="reviews"]')
        total_reviews.append(name.text.replace("/n","new line"))
    except:
        total_reviews.append('-')
       
    try:
        name=driver.find_element_by_xpath('//div[@class="keyword"]/span')
        overall_reviews.append(name.text.replace("/n","new line"))
    except:
        overall_reviews.append('-')    

    try:
        name=driver.find_element_by_xpath('//ul[@class="facility-sections"]')
        facilities.append(name.text.replace("/n","new line"))
    except:
        facilities.append('-')  
        
    try:
        name=driver.find_element_by_xpath('//div[@class="content collapse-content"]')
        property_description.append(name.text.replace("/n","new line"))
    except:
        property_description.append('-')  
```

```python
len(hostel_name),len(distance_from_city_centre),len(ratings),len(total_reviews),len(overall_reviews),len(privates_from_price),len(dorms_from_price),len(facilities),len(property_description)
```

```python
Hostels=pd.DataFrame()
Hostels['hostel_name']= hostel_name
Hostels['distance_from_city_centre']= distance_from_city_centre
Hostels['ratings']= ratings
Hostels['total_reviews']= total_reviews
Hostels['overall_reviews']= overall_reviews
Hostels['privates_from_price']= privates_from_price
Hostels['dorms_from_price']= dorms_from_price
Hostels['facilities']= facilities
Hostels['property_description']= property_description
        
Hostels
```
